# **outlier detection and handling**

```python
df.info()
```

```python
#define function to calculate IQR outliers
def detect_iqr_outliers(column):
  q1=column.quantile(0.25)
  q3=column.quantile(0.75)
  iqr=q3-q1
  lower_bound=q1-1.5*iqr
  upper_bound=q3+1.5*iqr
  return(column<lower_bound)| (column>upper_bound)

outliers_iqr=df[["size","total_sqft","bath","balcony","price"]].apply(detect_iqr_outliers)

print("Outliers detected uding IQR:")
print(outliers_iqr.sum())
```

```python
#filter out the rows with any outliers
df=df[~(outliers_iqr.any(axis=1))]
print("Data after removing outliers(IQR):")
print(df)
```

```python
#set the plot size for all plots
plt.figure(figsize=(8,10))

#create a boxplot
sns.boxplot(x="area_type",y="price",data=df,palette="Set2")
plt.title("Price Distribution by area_type status")
plt.xlabel("area_type status")
plt.ylabel("price")

#Adjust the layout to ensure the plots fit nicely
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import zscore

#compute zscores
z_scores=df[["size","total_sqft","bath","balcony","price"]].apply(zscore)

#define threshold
threshold=3

#identify outliers
outliers_z=(z_scores.abs()>threshold)

#view outliers
print("Outliers detected using Z-scores:")
print(outliers_z.sum())#count of outliers per column
```
